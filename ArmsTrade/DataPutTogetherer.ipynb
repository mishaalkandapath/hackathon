```python
allfiles = os.listdir('C:\\Users\\LiamDeaton\\Documents\\DATAProj\\ArmsTrade\\data')
for i in range(len(allfiles)):
    allfiles[i] = 'C:\\Users\\LiamDeaton\\Documents\\DATAProj\\ArmsTrade\\data\\' + allfiles[i]
compiled_df = []
for file in allfiles:
    df = pd.read_csv(file, skiprows=9)
    df.columns = df.iloc[0]
    with open(file, "r") as f:
        country = f.readline().strip().split()[5]
    df = df.drop(index=0)
    df.insert(0, 'exporter', country)
    display(df)
    compiled_df.append(df)
final = pd.concat(compiled_df, axis=0)
final.to_csv('final.csv')
```
